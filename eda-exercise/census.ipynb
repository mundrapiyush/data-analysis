```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
census = pd.read_csv('../datasets/census_dataset.csv')
census.AGE_GROUP.unique()
```

```python
females = census[(census.AGE_GROUP == '20-24') & (census.AREA_NAME == 'INDIA') & (census.AREA_TYPE == 'Total')]
females.ILIT_FEMALES / females.FEMALES
```

```python
pd.options.mode.copy_on_write = True
literacy = census[(census.AREA_NAME == 'INDIA') & (census.AREA_TYPE == 'Total')]
literacy_filtered = literacy[['AGE_GROUP','TOT','LIT_TOT']]
literacy_filtered['PCT'] = (literacy_filtered['LIT_TOT'] / literacy_filtered['TOT']) * 100
```

```python
sns.barplot(x = 'AGE_GROUP', y = 'PCT', data = literacy_filtered)
plt.xticks(rotation=90)
plt.show()
```

```python
census_state = census[census.AREA_NAME != 'INDIA']
census_state = census_state[['AREA_NAME','FEMALES','LIT_FEMALES']]
census_state['PCT'] = census_state['LIT_FEMALES'] / census_state['FEMALES'] * 100
```

```python
sns.barplot(x = 'AREA_NAME', y = 'PCT', data = census_state)
plt.xticks(rotation=90)
plt.show()
```
