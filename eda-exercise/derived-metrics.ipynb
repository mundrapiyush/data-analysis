```python
import pandas as pd
import numpy as np
cust_rating = pd.read_csv('https://media-doselect.s3.amazonaws.com/generic/GvOkzy5MLdy9zJ8vXMArwpRvN/rating_final.csv')
cust_rating.info()
cust_rating.head()
cust_rating['avg_rating'] = round((cust_rating.rating + cust_rating.food_rating + cust_rating.service_rating) / 3)
cust_rating
```

```python
odi_score = pd.read_csv('../datasets/odi-batting.csv')
odi_score['Centuries'] = odi_score.Runs >= 100
odi_score_by_centuries = odi_score.groupby(by='Player', axis=0).sum('Centuries')
odi_score_by_centuries.sort_values(by='Centuries', ascending=False)
```

```python
odi_score['RunRate'] = (odi_score.Runs / odi_score.Balls) * 100
odi_score[odi_score.Centuries > 0].sort_values(by='RunRate', ascending=False)
```

```python
odi_score['Year'] =  odi_score.MatchDate.str.split('-').str[2]
```

```python
odi_score_indians = odi_score[odi_score.Country == 'India']
```

```python
odi_score_indians_by_year = odi_score_indians.groupby(by=['Year','Player']).sum('Centuries').sort_values(by='Centuries', ascending=False)
```
